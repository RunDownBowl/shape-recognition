```python
import numpy as np
import matplotlib.pyplot as plt

from keras import Sequential
from keras.layers import Flatten, Dense, Dropout
from keras.utils import to_categorical
from keras.utils import image_dataset_from_directory

from PIL import Image
import os
```

```python
datasets_dir = "C:\\Users\\Erfan\\Desktop\\Code\\Become X\\session 1\\HW\\shape rec\\datasets\\train"

class_dirs = [os.path.join(datasets_dir, class_name) for class_name in os.listdir(datasets_dir) if os.path.isdir(os.path.join(datasets_dir, class_name))]

[print(f'Class: {os.path.basename(class_dir)}, Image: {os.listdir(class_dir)[0]}, Size: {Image.open(os.path.join(class_dir, os.listdir(class_dir)[0])).size}') for class_dir in class_dirs]
```

```python
datasets_dir_test = "C:\\Users\\Erfan\\Desktop\\Code\\Become X\\session 1\\HW\\shape rec\\datasets\\test"
```

```python
datasets_train = image_dataset_from_directory(datasets_dir, labels='inferred', label_mode='categorical', batch_size=32, image_size=(200,200), color_mode='grayscale', shuffle=True, seed=4)
```

```python
datasets_test = image_dataset_from_directory(datasets_dir_test, labels='inferred', label_mode='categorical', batch_size=1, image_size=(200,200), color_mode='grayscale', shuffle=False)
```

```python
def dataset_to_numpy(dataset):
    images = []
    labels = []
    for image_batch, label_batch in dataset:
        images.append(image_batch.numpy())
        labels.append(label_batch.numpy())
    return np.concatenate(images), np.concatenate(labels)
```

```python
xtrain, ytrain = dataset_to_numpy(datasets_train)
xtest, ytest = dataset_to_numpy(datasets_test)
```

```python
print(xtrain.shape)
print(ytrain.shape)
```

```python
xtrain = xtrain.reshape((xtrain.shape[0], 200*200))
xtrain = xtrain.astype('float32') / 255

xtest = xtest.reshape((xtest.shape[0], 200*200))
xtest = xtest.astype('float32') / 255
```

```python
model = Sequential()

model.add(Dense(100, 'relu', input_shape=(xtrain.shape[1],)))
#model.add(Dropout(0.5))
model.add(Dense(100, 'relu'))
#model.add(Dropout(0.5))
model.add(Dense(100, 'tanh'))
model.add(Dense(4, 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='MSE', metrics='accuracy')
```

```python
train_result = model.fit(xtrain, ytrain, epochs=15, batch_size=32, validation_split=0.2)
```

```python
t_loss, t_acc = model.evaluate(xtest, ytest)
print(t_loss)
print(t_acc)
```
